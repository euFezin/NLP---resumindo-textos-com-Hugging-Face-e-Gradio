```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
nome_modelo = "facebook/nllb-200-distilled-600M"

tokenizer_traducao = AutoTokenizer.from_pretrained(nome_modelo)
modelo_traducao = AutoModelForSeq2SeqLM.from_pretrained(nome_modelo)
```

```python
def traducao_eng_por(texto):
    tokenizer_traducao.src_lang = "eng_Latn"
    entrada = tokenizer_traducao(texto, return_tensors = "pt")
    bos_id = tokenizer_traducao.convert_tokens_to_ids("por_Latn")
    
    saida = modelo_traducao.generate(**entrada, forced_bos_token_id=bos_id, max_length=400)
    texto_traduzido = tokenizer_traducao.batch_decode(saida, skip_special_tokens=True)[0]
    return texto_traduzido
```

```python
app = gr.Interface(
    fn = traducao_eng_por,
    inputs = gr.Textbox(lines = 5, label = "Texto Original"),
    outputs = gr.Textbox(lines = 5, label = "Texto Traduzido"),
    title = "Tradutor de texto",
    description = "Envie o texto em português para traduzir para o Inglês."
)

app.launch()
```

```python
nome_modelo_traducao = 'facebook/nllb-200-distilled-600M'
nome_modelo_resumo = 'facebook/bart-large-cnn'

tokenizer_traducao = AutoTokenizer.from_pretrained(nome_modelo_traducao)
modelo_traducao = AutoModelForSeq2SeqLM.from_pretrained(nome_modelo_traducao)

tokenizer_resumo = AutoTokenizer.from_pretrained(nome_modelo_resumo)
modelo_resumo = AutoModelForSeq2SeqLM.from_pretrained(nome_modelo_resumo)


def traduzir(texto, idioma_origem, idioma_destino, max_length=512):
    tokenizer_traducao.src_lang = idioma_origem
    entrada = tokenizer_traducao(texto, return_tensors="pt", truncation=True)
    bos_id = tokenizer_traducao.convert_tokens_to_ids(idioma_destino)
    saida = modelo_traducao.generate(**entrada, forced_bos_token_id=bos_id, max_length=max_length)
    return tokenizer_traducao.batch_decode(saida, skip_special_tokens=True)[0]


def resumir(texto, max_length=200, min_length=100):
    entrada = tokenizer_resumo(texto, return_tensors="pt", truncation=True)
    saida = modelo_resumo.generate(**entrada, max_length=max_length, min_length=min_length)
    return tokenizer_resumo.batch_decode(saida, skip_special_tokens=True)[0]


def traduzir_resumir_texto(texto, ling_origem, ling_final):
    idioma_origem = codigos_linguagens[ling_origem]
    idioma_final = codigos_linguagens[ling_final]

    if idioma_origem == "eng_Latn":
        resumo = resumir(texto)
    else:
        texto_traduzido_en = traduzir(texto, idioma_origem, "eng_Latn")
        resumo = resumir(texto_traduzido_en)

    texto_final = traduzir(resumo, "eng_Latn", idioma_final)
    return texto_final
```
