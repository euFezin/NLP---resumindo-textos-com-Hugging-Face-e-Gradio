```python
from transformers import (
    pipeline,
    AutoTokenizer,
    AutoModelForSeq2SeqLM
)
```

```python
def carregar_modelo_trad(modelo, lingua_origem, lingua_destino):

    tokenizer = AutoTokenizer.from_pretrained(modelo)
    model = AutoModelForSeq2SeqLM.from_pretrained(modelo)

    tradutor = pipeline(
        task="translation",
        model=model,
        tokenizer=tokenizer,
        src_lang=lingua_origem,
        tgt_lang=lingua_destino
    )

    return tradutor
```

```python
modelo = "facebook/nllb-200-distilled-600M"

tradutor = carregar_modelo_trad(
    modelo,
    "eng_Latn",
    "por_Latn"
)

texto = """
Classification is one of the fundamental tasks in machine learning, where the
goal is to predict the category or class of a sample based on its features.
This technique is widely used in various applications such as image recognition,
medical diagnosis, spam filtering, sentiment analysis, and many others.
"""

resultado = tradutor(texto, max_length=200)

print(resultado[0]["translation_text"])
```

```python
def traduzir_texto(texto, lingua_origem, lingua_destino):

    tradutor = carregar_modelo_trad(
        "facebook/nllb-200-distilled-600M",
        lingua_origem,
        lingua_destino
    )

    traducao = tradutor(texto, max_length=400)

    return traducao[0]["translation_text"]
```

```python
def carregar_modelos(modelo_resumo,
                     modelo_traducao,
                     lingua_origem,
                     lingua_destino):

    tokenizer_resumo = AutoTokenizer.from_pretrained(modelo_resumo)
    model_resumo = AutoModelForSeq2SeqLM.from_pretrained(modelo_resumo)

    resumidor = pipeline(
        "summarization",
        model=model_resumo,
        tokenizer=tokenizer_resumo
    )

    tokenizer_trad = AutoTokenizer.from_pretrained(modelo_traducao)
    model_trad = AutoModelForSeq2SeqLM.from_pretrained(modelo_traducao)

    tradutor = pipeline(
        "translation",
        model=model_trad,
        tokenizer=tokenizer_trad,
        src_lang=lingua_origem,
        tgt_lang=lingua_destino
    )

    return resumidor, tradutor
```

```python
def resumir_traduzir(texto):

    resumidor, tradutor = carregar_modelos(
        "facebook/bart-large-cnn",
        "facebook/nllb-200-distilled-600M",
        "eng_Latn",
        "por_Latn"
    )

    resumo = resumidor(
        texto,
        max_length=200,
        min_length=100,
        do_sample=False
    )

    resumo = resumo[0]["summary_text"]

    traducao = tradutor(
        resumo,
        max_length=400
    )

    return traducao[0]["translation_text"]
```
